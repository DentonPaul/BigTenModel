```python
import requests
import pandas as pd
import numpy as np
```

```python
df_games = pd.read_csv('data/game_id_all_games.csv', index_col = "Unnamed: 0")
df_games['link'] = ''
df_games
```

```python
for index, row in df_games.iterrows():
    
    opponent = row.Opponent.lower()
    if row.Opponent.lower() == 'purdue-fort wayne':
        opponent = 'ipfw'
    
    if row.Opponent.lower() == 'miami (fl)':
        opponent == 'miami-fl'
        
    if row.Opponent.lower() == 'cal state long beach':
        opponent = 'long-beach-state'
        
        
    team_list = [row.TEAM.lower(), opponent]
    if ' ' in row.TEAM:
        team_list.append(row.TEAM.replace(' ', '-').lower())
        
    if ' ' in row.Opponent:
        team_list.append(row.Opponent.replace(' ', '-').lower())
    
    int_list = list(range(0, 30))
    
    date = str(row.Date)
    
    link = get_working_link(date, team_list, int_list)
    
    if type(link) == int:
        print(team_list, date)
        
    df_games.at[index, 'link'] = str(link)
```

```python
def get_working_link(date, team_list, int_list):
    
    # no int
    for team in team_list:
        resp = requests.get(f'https://www.sports-reference.com/cbb/boxscores/{date}-{team}.html')
        if resp.status_code == 200:
            return f'https://www.sports-reference.com/cbb/boxscores/{date}-{team}.html'
        
    # with int
    for team in team_list:
        for i in int_list:
            resp = requests.get(f'https://www.sports-reference.com/cbb/boxscores/{date}-{i}-{team}.html')
            if resp.status_code == 200:
                return f'https://www.sports-reference.com/cbb/boxscores/{date}-{i}-{team}.html'
            
    return -1
```
