```python
g_id = list(df.index)

def get_game_id(row):
    
    team = row.TEAM
    opp = row.Opponent
    
    df_sub = df[(df.Date==row.Date) & (df.TEAM==opp) & (df.Opponent == team)]
    
    if df_sub.shape[0] == 1:
        return row.index
    return 0
        
```
