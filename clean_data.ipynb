```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/bigten_data_joined.csv')
df.head()
```

```python
df['open_spread'] = df.open_spread.abs()
df['close_spread'] = df.close_spread.abs()

def spread(row):
    if not np.isnan(row.close_spread):
        return abs(row.close_spread)
    else:
        return abs(row.open_spread)
    
def tot(row):
    if not np.isnan(row.close_tot):
        return abs(row.close_tot)
    else:
        return abs(row.open_tot)
    
df['spread'] = df.apply(lambda row: spread(row), axis=1)
df['tot'] = df.apply(lambda row: tot(row), axis=1)
```

```python
# df = df.dropna().reset_index(drop=True)
drop_columns_X = []
for col in list(df.columns):
    nulls = df[col].isnull().sum()
    if nulls > 8:
        drop_columns_X.append(col)
```

```python
game_df_columns = ['game_id', 
                   'Date', 
                   'season', 
                   'season_num', 
                   'Playoffs', 
                   'Home', 
                   'Home_PTS', 
                   'Away', 
                   'Away_PTS', 
                   'open_spread', 
                   'close_spread',
                   'open_tot', 
                   'close_tot', 
                   'Home_ML', 
                   'Away_ML',
                   'OT', 
                   'winner']

game_df = df[game_df_columns]
game_df.head()
```

```python
game_df
```

```python
y_columns = ['game_id', 'OT', 'Tot', 'Spread', 'winner']

y_df = df[y_columns]
y_df
```

```python
X_columns = [col for col in df.columns if col not in y_columns]
X_columns = [col for col in X_columns if col not in ['Date','Home','Home_PTS','Away','Away_PTS', 'season', 'season_num', 'open_spread', 'close_spread', 'open_tot', 'close_tot']]
X_columns = [col for col in X_columns if col not in drop_columns_X]
X_columns = ['game_id'] + X_columns
```

```python
X_df = df[X_columns]
X_df = X_df.dropna()
```

```python
for col in y_df.columns:
    if col == 'game_id':
        continue
    X_df[col] = X_df.game_id.map(dict(y_df.set_index('game_id')[col]))
# X_df['OT'] = X_df.game_id.map(dict(y_df.set_index('game_id')['winner']))
```
