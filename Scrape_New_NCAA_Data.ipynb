```python
import pandas as pd
import numpy as np
import requests
import datetime
import re
```

```python
# base_url = 'https://www.sports-reference.com/cbb/boxscores/'
base_url = 'https://www.sports-reference.com/cbb/schools/{0}/{1}-schedule.html'
```

```python
bigten_teams = {
    'indiana': 'Indiana', 
    'maryland': 'Maryland',
    'michigan': 'Michigan',
    'michigan-state': 'Michigan State',
    'ohio-state': 'Ohio State',
    'penn-state': 'Penn State',
    'rutgers': 'Rutgers',
    'illinois': 'Illinois',
    'iowa': 'Iowa',
    'minnesota': 'Minnesota',
    'nebraska': 'Nebraska',
    'northwestern': 'Northwestern',
    'purdue': 'Purdue',
    'wisconsin': 'Wisconsin'
}
```

```python
df_list = []
for team in list(bigten_teams.keys()):
    for year in range(2010, 2022):
        url = base_url.format(team, year)
        df = pd.read_html(url)[1]
        df = df[df.G != 'G']
        df['TEAM'] = team
        df['season_start'] = year
        df_list.append(df)
df_list
        
```

```python
df_all = pd.concat(df_list)
```

```python
df_all = df_all.rename(columns={'season_start': 'season_end'})
```

```python
cols = [x for x in df_all.columns if 'Unnamed' not in x]
cols = cols[:-1]
```

```python
df_all = df_all[cols]
```

```python
def get_W_L(row):
    if int(row['Tm']) > int(row['Opp']):
        return True
    else:
        return False
    
df_all['Won'] = df.apply(lambda row: get_W_L(row), axis=1)
df_all['diff'] = df_all['Tm'].astype(int) - df_all['Opp'].astype(int)
    
```

```python
def get_won(row):
    if row['diff'] > 0:
        return True
    else:
        return False
    
df_all['Won'] = df_all.apply(lambda row: get_won(row), axis=1)
df_all
```

```python
def get_Date(row):
    date_str = row['Date']
    datetime.datetime.strptime(date_str, '')
```

```python
df_all['Date'] = pd.to_datetime(df_all['Date'])
```

```python
def team_clean(row):
    team = row.Opponent
    r = re.findall('\(\d+\)', team)
    if len(r) == 0:
        return team.strip()
    return team.replace(r[0], '').strip()
```

```python
df_all['Opponent'] = df_all.apply(lambda row: team_clean(row), axis=1)
df_all
```
